```python
X=list(data.iloc[:,5])
Y=list(data.iloc[:,3])
plot.bar(X,Y,color='crimson')
plot.title("Top 10 Salary")
plot.xlabel("Location")
plot.ylabel("Salary")
plot.xticks(rotation=90)
```

This bar chart shows peoples in New Delhi are getting more salary than other location.

```python
plot.hist(data['Rating'],color="green")
plot.show()
```

This histogram shows more than 7000 peoples are under the category of getting their salary based on Ratings between 3-4.

```python
plot.scatter(data['Salary'],data['Rating'])
plot.xticks(color='red',fontweight='bold',fontsize='12',horizontalalignment='right')
plot.show()
```

using Scatter plot We are comparing the salary and rating which shows that the people who get the highest salary has given the highest rating. Through the this we can find that few highly paid Data Professionals who are paid the most has given the rating 5. We can understand that Salary plays a major role in deciding the rating.
